5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class ThreeDFigure:
    def get_volume(self):
        return 0

    def __add__(self, other):
        if isinstance(other, ThreeDFigure):
            return self.get_volume() + other.get_volume()
        return NotImplemented
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cylinder(ThreeDFigure):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def get_volume(self):
        import math
        return math.pi * (self.radius ** 2) * self.height

class Cube(ThreeDFigure):
    def __init__(self, side_length):
        self.side_length = side_length

    def get_volume(self):
        return self.side_length ** 3

cylinder = Cylinder(3, 5)
cube = Cube(4)
total_volume = cylinder + cube
print(total_volume)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, id):
        self.__balance = 0
        self.__id = id

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Balance cannot be negative.")

    @property
    def id(self):
        return self.__id

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        else:
            print("Withdrawal amount must be positive and less than or equal to the balance.")
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.__items = []

    def add_item(self, item):
        self.__items.append(item)
        print(f"Added {item.name} to the cart.")

    def remove_item(self, item_name):
        initial_count = len(self.__items)
        self.__items = [item for item in self.__items if item.name != item_name]
        if len(self.__items) < initial_count:
            print(f"Removed {item_name} from the cart.")
        else:
            print(f"{item_name} not found in the cart.")

    def display_items(self):
        if not self.__items:
            print("The shopping cart is empty.")
        else:
            print("Items in the shopping cart:")
            for item in self.__items:
                print(f"- {item.name}: ${item.price}")
```

Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vehicle:
    def __init__(self, brand, production_age):
        self.brand = brand
        self.production_age = production_age

class Engine:
    def __init__(self, type, size):
        self.type = type
        self.size = size

class Car(Vehicle, Engine):
    def __init__(self, brand, production_age, type, size):
        Vehicle.__init__(self, brand, production_age)
        Engine.__init__(self, type, size)

my_car = Car("Toyota", 5, "petrol", 2.0)
print(my_car.brand)
print(my_car.size)
print(Car.mro())
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekIterator:
    def __init__(self):
        self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration
```
